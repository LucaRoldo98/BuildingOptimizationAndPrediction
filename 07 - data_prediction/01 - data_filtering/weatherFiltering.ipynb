```python
import pandas as pd
from datetime import datetime
```

```python
filepath = "Torre_Pellice_from_2017-01-01_to_2020-12-31"
df = pd.read_csv(filepath)
df
```

```python
# Create year, month, day and hour columns
def getYear(row):
    return row["Date_Time"].split(" ")[0].split("-")[0]
def getMonth(row):
    return row["Date_Time"].split(" ")[0].split("-")[1]
def getDay(row):
    return row["Date_Time"].split(" ")[0].split("-")[2]
def getHour(row):
    return row["Date_Time"].split(" ")[1].split(":")[0]
df['Year'] = df.apply(lambda x: getYear(x), axis=1)
df['Month'] = df.apply(lambda x: getMonth(x), axis=1)
df['Day'] = df.apply(lambda x: getDay(x), axis=1)
df['Hour'] = df.apply(lambda x: getHour(x), axis=1)
# Select only the columns we're intereseted in 
df = df[["Year", "Month","Day", "Hour", "Temperature_C"]]
df
```

```python
# Group by year, month, day and hour and average the temperature
df = df.groupby(["Year", "Month","Day","Hour"], as_index=False).Temperature_C.mean()
df
```

```python
# Now we need to average temperatures over multiple years, therefore average over month, day and hour (excluding year)
df1 = df.groupby(["Month", "Day", "Hour"], as_index=False).Temperature_C.mean()
df1

```

```python
df2 = df.groupby(["Month", "Day", "Hour"], as_index=False).Temperature_C
#.Temperature_C.std()
df2 = df2.var()
df2
```

```python
# Finally, return do datetime as to have only 2 columns. 2020 is put as the year in order to have a datetime format, 
#  however the value doesn't come from 2020 but from the average over 2019, 2020 and 2021
def backToDatetime(row):
    return ("2020/" + str(row["Month"])+"/"+str(row["Day"])+" "+str(row["Hour"])+":00:00")
df1["Date_Time"] = df1.apply(lambda x: backToDatetime(x), axis=1)
# Select only the columns we want
df1 = df1[["Date_Time", "Temperature_C"]]
df1
```

```python
df3 = df1
df3["Variance"] = df2["Temperature_C"]
```

```python
# Save the output in a csv file 
df3.to_csv("averagedOverHourAndYearsTemperatures.csv")
```
